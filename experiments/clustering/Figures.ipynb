**1. Australia Electricity data**

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random

fontsizes = [8, 16, 24, 32]

def example_plot(ax, consumption):
    #ax.plot([1, 2])
    hours=range(24)
    ax.plot(hours, consumption, '-o')
    #ax.set_xlabel('x-label', fontsize=random.choice(fontsizes))
    #ax.set_ylabel('y-label', fontsize=random.choice(fontsizes))
    #ax.set_title('Title', fontsize=random.choice(fontsizes))

    
data = np.loadtxt("au_centroids.csv", delimiter=',')    

fig, axs= plt.subplots(nrows=4, ncols=4)
i = 0
for row in axs:
    for ax in row:
        if i<len(data):
            example_plot(ax,data[i])
            i = i+1
plt.tight_layout()

plt.show()
```
